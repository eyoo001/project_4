```python
def impute_ems(df, impute_date, col, stn):
    if stn==1:
        row_num = ps_w[(ps_w.loc[:,'Date']==impute_date) & (ps_w.loc[:,'Station']==1)].index[0]
        ps_w.loc[row_num, col] = ps_w[(ps_w['Date']==impute_date) & 
                                                 (ps_w['Station']==2)][col].iloc[0]
    else:
        row_num = ps_w[(ps_w.loc[:,'Date']==impute_date) & (ps_w.loc[:,'Station']==2)].index[0]
        ps_w.loc[row_num, col] = ps_w[(ps_w['Date']==impute_date) & 
                                                 (ps_w['Station']==1)][col].iloc[0]
```
