```python
import folium
from folium import plugins
import pandas as pd
from datetime import datetime, timedelta
```

```python
train = pd.read_csv('./data/train.csv')
```

```python
train.info()
```

```python
train.Date = train.Date.map(lambda x: datetime.strptime(x, '%Y-%m-%d'))
```

```python
map_me = folium.Map(location=[train.Latitude.mean(), train.Longitude.mean()], zoom_start=10)
```

```python
hotdf = train\
    .assign(month = lambda df: df.Date.map(lambda x: x.month), year = lambda df: df.Date.map(lambda x: x.year))\
    .groupby(['year', 'month', 'Latitude', 'Longitude'])\
    .WnvPresent\
    .sum()\
    .reset_index()
```

```python
hotlists = [[[raw.Latitude, raw.Longitude, raw.WnvPresent] for i, raw in hotdf.query(f'month=={row[1]} and year=={row[0]}').iterrows() if raw.WnvPresent>0] 
            for row in hotdf.set_index(['year', 'month']).index.unique()]
```

```python
labels = [f'{row[1]}/{row[0]}' for row in hotdf.set_index(['year', 'month']).index.unique()]
```

```python
hotmap = folium.plugins.HeatMapWithTime(hotlists, index=labels, radius=30)
```

```python
hotmap.add_to(map_me)
```
